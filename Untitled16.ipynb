Create a Bar Chart Using Matplotlib.
â€¢ Population of Countries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("population.csv")
```

```python
# Sort by population and select top 5
top5 = df.sort_values(by="Population", ascending=False).head(5)
```

```python
# Plot bar chart
plt.figure(figsize=(8,6))
plt.bar(top5["Country"], top5["Population"], color="skyblue", edgecolor="black")
```

```python
# Title and labels
plt.title("Population of Top 5 Countries", fontsize=14, weight="bold")
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population (in Billions)", fontsize=12)
```

```python
# Format y-axis to billions
plt.gca().set_yticks([0, 500_000_000, 1_000_000_000, 1_500_000_000])
plt.gca().set_yticklabels(["0", "0.5B", "1B", "1.5B"])
```

```python
# Show gridlines
plt.grid(axis="y", linestyle="--", alpha=0.6)
```

```python
# Display chart
plt.show()
```

Create a Bar Chart Using Matplotlib.
â€¢ Number of Movies Produced by Genres

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("movies_genre.csv")
```

```python
# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(df["Genre"], df["Movie_Count"], color="teal", edgecolor="black")
```

```python
# Title and labels
plt.title("Number of Movies Produced by Genre", fontsize=14, weight="bold")
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
```

```python
# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha="right")
```

```python
# Add gridlines
plt.grid(axis="y", linestyle="--", alpha=0.6)
```

```python
# Display chart
plt.tight_layout()
plt.show()
```
