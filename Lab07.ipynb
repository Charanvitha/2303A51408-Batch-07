**Create a Line Chart Using Matplotlib**


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv("daily_temperature.csv")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(weather["Day"], weather["Temperature"], marker='o', linestyle='-', color='b', label="Temperature")
```

```python
plt.title("Daily Temperature Variation Over a Month", fontsize=14)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Temperature (Â°C)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
```

```python
plt.show()
```

```python
hottest_day = weather.loc[weather["Temperature"].idxmax()]
coldest_day = weather.loc[weather["Temperature"].idxmin()]

print(f"Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}Â°C")
print(f"Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}Â°C")
```

**Create a Line Chart Using Matplotlib**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("sales_data.csv")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(sales["Quarter"], sales["Sales"], marker='o', linestyle='-', color='g', label="Sales")
```

```python
peak = sales.loc[sales["Sales"].idxmax()]
plt.scatter(peak["Quarter"], peak["Sales"], color='red', s=100, label="Peak Sales")
```

```python
plt.title("Quarterly Sales Performance", fontsize=14)
plt.xlabel("Quarter", fontsize=12)
plt.ylabel("Sales (in units)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
```

```python
plt.show()
print(f"Peak Sales: Quarter {peak['Quarter']} with {peak['Sales']} units")
```
